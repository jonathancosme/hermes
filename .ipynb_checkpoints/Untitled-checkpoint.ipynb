```python
import numpy as np
import string
import random
import pandas as pd
```

```python
from datetime import datetime
class PhoneObject():
    """
    phone object designed to mimic gathering lat and lon data, and sending it to a remote server
    """
    def __init__(self, send_fail_prob=0.2):
        super(PhoneObject, self).__init__()
        self.send_fail_prob = send_fail_prob
        self.phone_id = gen_random_id(5)
        self.phone_model = gen_random_phone_model()
        self.software_version = gen_random_software_version()
        self.data = []
        
    def read_and_note_cur_lat_lon(self):
        data = {}
        data['time'] = datetime.now()
        data['lat'] = np.random.uniform(low=-90., high=90., size=1)[0]
        data['lon'] = np.random.uniform(low=-180., high=180., size=1)[0]
        return data
    
    def try_to_send_data(self):
        self.data.append(self.read_and_note_cur_lat_lon())
        fail_lotto = np.random.uniform()
        if fail_lotto > self.send_fail_prob:
            
            send_data = {
                'phone_id': self.phone_id,
                'phone_model': self.phone_model,
                'software-version': self.software_version,
                'data': np.array(self.data),
                'sent_timestamp': datetime.now(),
            }
            
            self.data = []
            
            return send_data
        else:
            return None
            
            
        
    
```

```python
np.random.uniform()
```

```python
def gen_random_id(y):
    rand_id = ''.join(random.choice(string.ascii_uppercase) for x in range(y))
    rand_id += ''.join(np.random.choice(range(10), size=y).astype(str))
    return rand_id
```

```python
gen_random_id(5)
```

```python
all_phone_models = ['iphone', 'galaxy', 'pixel', 'htc', 'nokia']
def gen_random_phone_model():
    rand_phone_model = random.choice(all_phone_models) + '-'
    rand_phone_model += ''.join(np.random.choice(range(10)).astype(str))
    return rand_phone_model
```

```python
gen_random_phone_model()
```

```python
def gen_random_software_version():
    software_version = '.'.join(np.random.choice(range(10), size=3).astype(str))
    return software_version
```

```python
gen_random_software_version()
```

```python
aPhoneObject = PhoneObject()
print(aPhoneObject.phone_id)
print(aPhoneObject.phone_model)
print(aPhoneObject.software_version)
print(aPhoneObject.read_and_note_cur_lat_lon())
```

```python
data = []
for i in range(10):
    data.append(aPhoneObject.try_to_send_data())
data = [x for x in data if x != None]
```

```python
df = pd.DataFrame.from_records(data)
df
```
